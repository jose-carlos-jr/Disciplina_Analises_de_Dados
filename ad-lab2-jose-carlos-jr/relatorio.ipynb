```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Substitua o filtro com o nome da série que você deseja analisar
num_episodio = (
    pd.read_csv('dados/avaliacao_episodios_series_top.csv')
    .query("nome_serie == 'Death Note'")
)
num_episodio
```

## Popularidade dos episódios

### Qual a distribuição de votos que cada episódio teve?

Explore a popularidade de cada episódio analisando a quantidade de votos que cada um recebeu (`num_votos`). Exiba visualizações das distribuções e calcule estatísticas de centralidade e dispersão para esta variável.


```python
# INSIRA SEU CÓDIGO AQUI
plt.subplots(figsize=(10,4))
ax = sns.histplot(x='num_votos', data = num_episodio, color='green')
ax = ax.set_xlabel("Votos por Episódios"), ax.set_title("Dispersão dos votos")
plt.show()
```

```python
plt.subplots(figsize=(10,4))
ax = sns.boxplot(x='num_votos', data = num_episodio, color='green')
ax = ax.set_xlabel("Votos por Episódios"), ax.set_title("Centralidade dos votos")
plt.show()
```

```python
media = num_episodio['num_votos'].mean()
mediana = num_episodio['num_votos'].median()
moda = num_episodio['num_votos'].mode()

print(f"Média: {media} / Mediana: {mediana} / Moda: {float(moda[0])}, {float(moda[1])}")
```

### Qual o episódio mais popular (com mais votos)?

```python
# INSIRA SEU CÓDIGO AQUI
num_episodio.sort_values('num_votos', ascending=False).head(1)
```

### Qual o episódio menos popular (com menos votos)?

```python
# INSIRA SEU CÓDIGO AQUI
num_episodio.sort_values('num_votos', ascending=True).head(1)
```

### Qual a temporada mais popular?

```python
# INSIRA SEU CÓDIGO AQUI
temporada = num_episodio.groupby(['temporada'])['num_votos'].sum().sort_values(ascending=False).head(1)
x = temporada.reset_index()
x
```

### Qual a temporada menos popular?

```python
# INSIRA SEU CÓDIGO AQUI
temporada = num_episodio.groupby(['temporada'])['num_votos'].sum().sort_values(ascending=True).head(1)
x = temporada.reset_index()
x
```

## Relação entre variáveis

Analise a relação entre as variáveis referentes a cada episódio. Explore visualmente a relação entre elas e calcula estatísticas de correlação.

### Existe alguma relação entre a quantidade de votos recebida e a avaliação média dos episódios?

```python
# INSIRA SEU CÓDIGO AQUI
print(f"Correlação: {num_episodio.num_votos.corr(num_episodio.avaliacao_media)}")
sns.lmplot(num_episodio, x = 'num_votos', y = 'avaliacao_media')


pd.Series({
    'pearson': num_episodio.num_votos.corr(num_episodio.avaliacao_media, method='pearson'),
    'spearman': num_episodio.num_votos.corr(num_episodio.avaliacao_media, method='spearman'),
    'kendall': num_episodio.num_votos.corr(num_episodio.avaliacao_media, method='kendall'),
}, name='correlacao').to_frame()
```

### Existe alguma relação entre a quantidade de votos recebida e o número do episódio?

```python
# INSIRA SEU CÓDIGO AQUI
print(f"Correlação: {num_episodio.num_votos.corr(num_episodio.num_episodio)}")
sns.lmplot(num_episodio, x ='num_episodio', y = 'num_votos')

pd.Series({
    'pearson': num_episodio.num_votos.corr(num_episodio.num_episodio, method='pearson'),
    'spearman': num_episodio.num_votos.corr(num_episodio.num_episodio, method='spearman'),
    'kendall': num_episodio.num_votos.corr(num_episodio.num_episodio, method='kendall')
}, name='correlacao').to_frame()
```
